```python
def verdict(item):
    if re.search(r',  соответствует лицензионным требованиям', item):
        return 'meets_reqs'
    elif: re.search(r', не соответствует лицензионным требованиям', item):
        return 'dn_meet_reqs'   
    elif: re.search(r', Выявлены нарушения', item) or re.search(r' Выявлены.', item):
        return 'violations_found'
    elif: re.search(r', Нарушения, указанные в предписании Рособрнадзора, устранены', item):
        return 'violations_eliminated'
    elif: item == 'Объект проверки отсутствует':
            return 'missing'
    else:
        return item
```
