```python
# reading data and dictionary
raw_data = json.load(open("indicators/data_2018.json", 'r'))
raw_data_2017 = json.load(open("indicators/data_2017.json", 'r'))
raw_status = json.load(open("status/data.json", 'r'))
raw_dict = json.load(open("indicators/dict.json", 'r'))[0]
```

```python
# insert missing keys to all the dictionaries
max_key = max(max(d) for d in raw_data)
empty = dict.fromkeys(set().union(*raw_data), None)
raw_data_1 = [dict(empty, **d) for d in raw_data]

max_key = max(max(d) for d in raw_data_2017)
empty = dict.fromkeys(set().union(*raw_data_2017), None)
raw_data_1_2017 = [dict(empty, **d) for d in raw_data_2017]
```

```python
# preparing final dictionary
final_dict_2018 = collections.defaultdict(list)
for d in raw_data_1:
    for k, v in d.items(): 
        final_dict_2018[k].append(v)
        
final_dict_2017 = collections.defaultdict(list)
for d in raw_data_1_2017:
    for k, v in d.items(): 
        final_dict_2017[k].append(v)        
```

```python
df_raw = pd.DataFrame.from_dict(final_dict_2018)
df_raw_2017 = pd.DataFrame.from_dict(final_dict_2017)
```

```python
def num_format(x):
    if isinstance(x, str):
        comma_cnt = x.count(',')
        if comma_cnt == 1 and x.replace(' ', '').replace(',', '').replace('%', '').replace('+', '').replace('–', '').isdigit():
            x = x.replace(',', '.')
            x = x.replace(' ', '')
            x = x.replace('%', '')
            try:
                return float(x)
            except ValueError:
                return x
        elif comma_cnt == 0:
            x = x.replace('%', '')
            try:
                return float(x.replace(' ', ''))
            except ValueError:
                return x
        else:
            return x
    else:
        return x
```

```python
df_tmp = df_raw.applymap(num_format)
df = df_tmp.drop(df_tmp.columns[df_tmp.isna().all()], axis=1)

df_tmp_2017 = df_raw_2017.applymap(num_format)
df_2017 = df_tmp_2017.drop(df_tmp_2017.columns[df_tmp_2017.isna().all()], axis=1)
```
