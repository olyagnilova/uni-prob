```python
import json
import csv
import collections
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_columns', 500)
```

```python
def data_prep(path):
    raw_data = json.load(open(path, 'r'))
    # insert missing keys to all the dictionaries
    max_key = max(max(d) for d in raw_data)
    empty = dict.fromkeys(set().union(*raw_data), None)
    raw_data_1 = [dict(empty, **d) for d in raw_data]
    # preparing final dictionary
    final_dict = collections.defaultdict(list)
    for d in raw_data_1:
        for k, v in d.items(): 
            final_dict[k].append(v)
    df_raw = pd.DataFrame.from_dict(final_dict)
    return df_raw
```

```python
def num_format(x):
    if isinstance(x, str):
        comma_cnt = x.count(',')
        if comma_cnt == 1 and x.replace(' ', '').replace(',', '').replace('%', '').replace('+', '').replace('–', '').isdigit():
            x = x.replace(',', '.')
            x = x.replace(' ', '')
            x = x.replace('%', '')
            try:
                return float(x)
            except ValueError:
                return x
        elif comma_cnt == 0:
            x = x.replace('%', '')
            try:
                return float(x.replace(' ', ''))
            except ValueError:
                return x
        else:
            return x
    else:
        return x
```

```python
df_raw_2018 = data_prep("indicators/data_2018.json")
df_raw_2017 = data_prep("indicators/data_2017.json")
df_raw_2016 = data_prep("indicators/data_2016.json")

df_tmp_2018 = df_raw_2018.applymap(num_format)
df_2018 = df_tmp_2018.drop(df_tmp_2018.columns[df_tmp_2018.isna().all()], axis=1)

df_tmp_2017 = df_raw_2017.applymap(num_format)
df_2017 = df_tmp_2017.drop(df_tmp_2017.columns[df_tmp_2017.isna().all()], axis=1)

df_tmp_2016 = df_raw_2016.applymap(num_format)
df_2016 = df_tmp_2016.drop(df_tmp_2016.columns[df_tmp_2016.isna().all()], axis=1)
```
