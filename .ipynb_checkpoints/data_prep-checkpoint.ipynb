```python
import json
import csv
import collections
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_columns', 500)
```

```python
# reading data and dictionary
raw_data = json.load(open("indicators/data_2018.json", 'r'))
raw_status = json.load(open("status/data.json", 'r'))
raw_dict = json.load(open("indicators/dict.json", 'r'))[0]
```

```python
# save dict
# with open('indicators_dict.csv', 'w') as f:
#     w = csv.DictWriter(f, raw_dict.keys())
#     w.writeheader()
#     w.writerow(raw_dict)
```

```python
# insert missing keys to all the dictionaries
max_key = max(max(d) for d in raw_data)
empty = dict.fromkeys(set().union(*raw_data), None)
raw_data_1 = [dict(empty, **d) for d in raw_data]
```

```python
# preparing final dictionary
final_dict = collections.defaultdict(list)
for d in raw_data_1:
    for k, v in d.items(): 
        final_dict[k].append(v)
```

```python
df_raw = pd.DataFrame.from_dict(final_dict)
df_raw.head()
```

```python
def num_format(x):
    if isinstance(x, str):
        comma_cnt = x.count(',')
        if comma_cnt == 1 and not x.replace(' ', '').replace(',', '').isalpha():
            x = x.replace(',', '.')
            x = x.replace(' ', '')
            x = x.replace('%', '')
            try:
                return float(x)
            except ValueError:
                return x
        elif comma_cnt == 0:
            x = x.replace('%', '')
            try:
                return float(x.replace(' ', ''))
            except ValueError:
                return x
        else:
            return x
    else:
        return x
```

```python
df_tmp = df_raw.applymap(num_format)
df = df_tmp.drop(df_tmp.columns[df_tmp.isna().all()], axis=1)
```
